## Face Detection
#### Import open cv and Assign HaarCascade

```python
import cv2
faceCascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

#### Read Image and Convert to GrayScale

```python
image = cv2.imread("a.jpg")
```

```python
from matplotlib import pyplot as plt
plt.imshow(image)
```

```python
RGB_img = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```

```python
GR_img = cv2.cvtColor(RGB_img,cv2.COLOR_BGR2GRAY)
plt.imshow(GR_img)
```

#### Detect Coordinations of Faces

```python
faces = faceCascade.detectMultiScale(GR_img, scaleFactor = 1.1, minNeighbors = 5, minSize = (30,30))
print(faces)
```

```python
image = cv2.imread("a.jpg")
for(x,y,w,h) in faces:
    cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,0,),5)
RGB_img = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```

#### Crop Faces

```python
ims = []
for (x,y,w,h) in faces:
    crop = RGB_img[y:y+h,x:x+w]
    ims.append(crop)
```
